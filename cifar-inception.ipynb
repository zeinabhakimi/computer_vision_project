## Inception Model and CIFAR10

Here you can find the example of loading a CIFAR10 image set and feeding it through Pre-trained Inception-v3. 


```python
from IPython.display import Image, display
from p2src.cifar10 import *
images, img_labels = load_images('./cifar-10-batches-py/', 'test_batch')
```

The actual image in CIFAR10 is just a 32x32 (RGB) image:

```python
PIL.Image.fromarray(images[70])
```

However, we have intercepted the output of Resize layer in the inception to show the effect of this layer and how it is working. The following code create an instance of inception model with pre-trained network (graph):

```python
from p2src import inception
model = inception.Inception('inception', 'classify_image_graph_def.pb')
```

The code below shows how to capture the resized image (right before feeding it through Inception's multiple layers. As you can see, Inception automatically resize any image to its default configuration 299x299 using bilinear interpolation.

```python
resized_image = model.get_resized_image(image=images[70])
PIL.Image.fromarray(resized_image)
```
