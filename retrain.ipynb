## Retraining Inception-v3 Network

In this method we have added two new layer (softmax and fully connected) to our pre-trained graph. We added the right operations to the graph, along with some variables to hold the weights, and then sets up all the gradients for the backward pass. Also, we have added a evaluation step to measure the accuracy of our results. The code for this step can be found in the [retrain_classify.py](https://github.com/MortezaRamezani/s17-cse586-p2/blob/master/p2src/retrain_classify.py). Most of the code for this method has been adopted from Tensorflow retrain examples. The complete references list can be found [here](https://github.com/MortezaRamezani/s17-cse586-p2/blob/master/p2src/LICENSE)

```python
import numpy as np
import matplotlib.pyplot as plt
```

Loading the results to plot the figures:

```python
training_results = np.load('./results/training_accuracy.npy')
test_result = np.load('./results/test_accuracy.npy')
```

The training accuracy shows what percent of the images used in the current training set were labeled with the correct classs. Here you can see the training accuracy during training steps:

```python
t = np.arange(0,training_results.shape[1],1)
plt.figure(figsize=(20, 5))
plt.plot(t, training_results[0] * 100)
plt.ylabel('Train Accuray (%)', fontsize=16)
plt.xlabel('Step', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

The following graph shows the "Cross Entropy" durig training steps.  Cross entropy is a loss function which shows us how well the learning process is progressing. The training goal is to minimize the loss as much as possible.

```python
t = np.arange(0,training_results.shape[1],1)
plt.figure(figsize=(20, 5))
plt.plot(t, training_results[1] * 100, '-')
plt.ylabel('Cross Entropy', fontsize=16)
plt.xlabel('Step', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

The validation accuracy is the precision on a randomly-selected group of images from a different set. After each epoch, we have used images from validation image set to measure the perfromance of network.

```python
t = np.arange(0,training_results.shape[1],1)
plt.figure(figsize=(20, 5))
plt.plot(t, training_results[2] * 100, '-')
plt.ylabel('Validation Accuracy (%)', fontsize=16)
plt.xlabel('Step', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

Finally, we have used a test image set, which has been kept seperate from training and validation image sets. The final accuracy for the test data was:

```python
print('%.2f%%' % (np.load('./results/test_accuracy.npy') * 100))
```
